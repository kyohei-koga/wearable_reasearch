### データ内容
- 全76社
- すべての企業について特許データとの対応付けは、時間が掛かり過ぎる。出展数の多い企業に注目するのが良いか。
- 企業名に表記ゆれあり。例）AppleとApple computer
- 輸入販売を行っている企業がmanufakuturerになっている場合がある。例）株式会社フューチャーズ

### 企業特徴
- 中国企業が多い（STK,Flylintech,Easyday,shenzen wave）

# ２．レビュー件数が多い企業

```python
data.sort_values("num_of_rate",ascending=False).head(20)
```

- Jawboneが１位
- レビュー件数(num_of_rate)と製造者(manufacturer)が同じ場合、同じ型の製品でデザイン違い（色違い等）の可能性大。そのため、レビュー内容は同じ。
- 中国企業もちらほら（huawei,STK,MisSmart,shenzhen wave）
- 上位はJawboneとFitbitが独占しているように見えるが、レビュー件数が同じ件数が並んでいるため、デザイン違いの可能性大

# ３．レビュー点数が高い企業

### レビュー件数10件以上

```python
data_10 = data[data["num_of_rate"]>=10].sort_values("average_rate",ascending=False)
data_10.head(10)
```

- GarminはGPS機器メーカ、山用品も強い
- UZOUは中国の情報機器専門ベンチャー
- Lendooは中華企業
- Wahoo Fitnessは多分米国企業。フィットネス製品とセンサーの融合が売り
- misfitはフォッシルグループのブランド。フォッシルは米国のアパレル企業

### レビュー件数20件以上

```python
data_20 = data[data["num_of_rate"]>=20].sort_values("average_rate",ascending=False)
data_20.head(10)
```

- 10件位上と変化なし

### レビュー件数30件以上

```python
data_30 = data[data["num_of_rate"]>=30].sort_values("average_rate",ascending=False)
data_30.head(10)
```

- Jscottonは中国企業
- IWは中国企業
