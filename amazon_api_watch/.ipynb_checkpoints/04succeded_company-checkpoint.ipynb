# 分類：成功企業
- 上市している企業
アマゾンデータに該当する企業すべて
- レビュー件数が多い製品を製造している企業
- レビュー点数が高い製品を製造している企業
- セールスランクが高い表品を製造している企業

```python
import pandas as pd
data = pd.read_csv("api_data_smart_watches.csv")
print data.columns
```

# １．上市している企業

```python
print("10製品以上\n")
data.manufacturer.value_counts()[:11]
```

### データ内容
- 全229社
- すべての企業について特許データとの対応付けは、時間が掛かり過ぎる。出展数の多い企業に注目するのが良いか。
- 企業名に表記ゆれあり。例）AppleとApple computer
- 輸入販売を行っている企業がmanufakuturerになっている場合がある。例）株式会社フューチャーズ

### 企業特徴
- 中国企業が多い（STK,Flylintech,Easyday,shenzen wave）

# ２．レビュー件数が多い企業

```python
data.sort_values("num_of_rate",ascending=False).head(20)
```

- Huaweiが１位
- レビュー件数(num_of_rate)と製造者(manufacturer)が同じ場合、同じ型の製品でデザイン違い（色違い等）の可能性大。そのため、レビュー内容は同じ。
- 中国企業もちらほら（huawei,STK,MisSmart,shenzhen wave）
- ネット販売専門はU-Watch(中国企業)の販売を行っている。
- withingsは、見た目は腕時計だが、機能はリストバンド型の範疇（アクティビティトラッカー）

# ３．レビュー点数が高い企業

### レビュー件数10件以上

```python
data_10 = data[data["num_of_rate"]>=10].sort_values("average_rate",ascending=False)
data_10.head(10)
```

- by Galaxyはサムスンの製品
- EloBethはサムスンスマートウォッチのアクセサリ→ノイズ
- TOP10はペブルとサムスンの製品のみ
- TOP20では、Asus,motorola,huawei,uzou(中国)がある。日本勢なし

### レビュー件数20件以上

```python
data_20 = data[data["num_of_rate"]>=20].sort_values("average_rate",ascending=False)
data_20.head(10)
```

- 20位以内からsamusungが外れた
- 20位以内にソニーあり

### レビュー件数30件以上

```python
data_30 = data[data["num_of_rate"]>=30].sort_values("average_rate",ascending=False)
data_30.head(10)
```

- 20位以内はpebbleとmotorolaのみ
- appleは出てこない。amazon経由での購入が少ない？

# ４．セールスランクの高い企業
- セールスランクはカテゴリー内のランキングで、本データではエレクトロニクス全体だと思われます。

```python
data.sort_values("rank").head(20)
```

- 1位と2位はスマートウォッチではない。ノイズ。
- レビュー件数がついていない製品は、発売されて間もない可能性あり。例：フィットビット。
- 日本企業はソニーだけ
- ランキングのカテゴリの確認必要

# ５．まとめ
- ノイズや表記ゆれの整理は必要
- 同じ製品のデザイン違いは重複したデータとして扱うのが良い
- いずれの選択基準でも、大体の顔ぶれは同じ
- 中国企業が結構いる。とにかく価格が安い、１桁違う。（一般：万円台、中国：千円台）
- 日本企業はソニーぐらい
- アップルウォッチはアマゾンを経由しないで購入されてい無いためか、情報が少ない

# ４．その他
- リストバンド型と考えられる製品も混ざっている。
- 抽出する成功企業は特許権数からして3〜4ぐらいが調度良いと思われる。
- rankを用いた抽出方法（上位（ウェアラブル）OR　1万位以内（カテゴリ全体））
